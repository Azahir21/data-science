```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn import cluster
```

```python
dataset = pd.read_csv('milk.csv')
display(dataset.tail())
```

```python
data = dataset.loc[:,['pH', 'Temprature', 'Taste', 'Odor', 'Turbidity', 'Colour']]
display(data.tail())
```

```python
sc = MinMaxScaler(feature_range=(0, 1))
data = sc.fit_transform(data)
print('Data setelah dinormalisasi: \n', data)
```

```python
clustering = KMeans(n_clusters=3, init='random', n_init = 3)
clusters = clustering.fit_predict(data)

print('Hasil clustering ke-10: \n', clusters)
print('\nSSE: ', clustering.inertia_,)
```

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage='single')
clusters = clustering.fit_predict(data)

print('Hasil clustering single: ', clusters)
```

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage='average')
clusters = clustering.fit_predict(data)

print('Hasil clustering average: ', clusters)
```

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage='complete')
clusters = clustering.fit_predict(data)

print('Hasil clustering complete: ', clusters)
```
