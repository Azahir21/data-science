```python
# Import Libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np

# Baca Dataset
dataset = pd.read_csv('milk.csv')
display(dataset.tail())

# Masukkan training data (Semua dari dataset)
train_data = np.array(dataset)[:,0:-1]
train_label = np.array(dataset)[:,-1]
```

```python
# Normalisasi Min Max
sc = MinMaxScaler(feature_range = (0,1))

data = sc.fit_transform(train_data)
print(data)
```

```python
# Deklarasi fungsi kNN, k = 3
kNN = KNeighborsClassifier(n_neighbors=3, weights='distance')
# Implementasi kNN ke training data
kNN.fit(train_data, train_label)
```

```python
# Input data
pH = input("pH: ")
Temp = input("Temp: ")
Taste = input("Taste: ")
Odor = input("Odor: ")
Fat = input("Fat: ")
Turbidity = input("Turbidity: ")
Colour = input("Colour: ")

test_data = np.array([float(pH), float(Temp), int(Taste), int(Odor), int(Fat), int(Turbidity), int(Colour)])
print(test_data)
```

```python
# Reshape data agar sesuai
test_data = np.reshape(test_data,(1,-1))
print(test_data)
```

```python
# Predict label
hasil = kNN.predict(test_data)

# Output
print("Hasil dari k-NN: ", hasil)
```
