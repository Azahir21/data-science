```python
# Import Libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

# Baca Dataset Training
training_dataset = pd.read_csv('milk_training.csv')
display(training_dataset.tail())

# Masukkan training data (Semua dari training_dataset)
train_data = np.array(training_dataset)[:,0:-1]
train_label = np.array(training_dataset)[:,-1]
```

```python
# Normalisasi Min Max
sc = MinMaxScaler(feature_range = (0,1))

train_data = sc.fit_transform(train_data)
print(train_data)
```

```python
# Baca Dataset Testing
testing_dataset = pd.read_csv('milk_testing.csv')
display(testing_dataset)

# Masukkan testing data (Semua dari testing_dataset)
test_data = np.array(testing_dataset)[:,0:-1]
test_label = np.array(testing_dataset)[:,-1]
```

```python
# Normalisasi Min Max Testing Data
sc = MinMaxScaler(feature_range = (0,1))

test_data = sc.fit_transform(test_data)
print(test_data)
```

```python
# Deklarasi fungsi kNN, k = 3
kNN = KNeighborsClassifier(n_neighbors=3, weights='distance')
# Implementasi kNN ke training data
kNN.fit(train_data, train_label)
```

```python
# Predict label
hasil = kNN.predict(test_data)

# Nilai akurasi
Score = accuracy_score(test_label, hasil)
print('Nilai akurasi: ',Score)

# Output
print("\nHasil dari k-NN: ", hasil)
```
