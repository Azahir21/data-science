```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('titanic.csv')
print(dataset)
```

```python
rows, cols = dataset.shape
print("Jumlah baris", rows)
print("Jumlah kolom", cols)
```

```python
data = dataset.loc[:, ['Age', 'Fare']]
print(data)
```

```python
kelas = dataset.loc[:, ['Survived']]
print(kelas)
```

```python
dataset = dataset.fillna(dataset.groupby('Survived').transform('mean'))
print(dataset)
```

```python
# Normalisasi Min-Max (0-1) untuk kolom 'Age' dan 'Fare'
min_age = dataset['Age'].min()
max_age = dataset['Age'].max()
min_fare = dataset['Fare'].min()
max_fare = dataset['Fare'].max()

dataset['Age_MinMax'] = (dataset['Age'] - min_age) / (max_age - min_age)
dataset['Fare_MinMax'] = (dataset['Fare'] - min_fare) / (max_fare - min_fare)

print(dataset[['Age_MinMax', 'Fare_MinMax', 'Age', 'Fare']])
```

```python
# Normalisasi Z-Score untuk kolom 'Age' dan 'Fare'
mean_age = dataset['Age'].mean()
std_age = dataset['Age'].std()
mean_fare = dataset['Fare'].mean()
std_fare = dataset['Fare'].std()

dataset['Age_ZScore'] = (dataset['Age'] - mean_age) / std_age
dataset['Fare_ZScore'] = (dataset['Fare'] - mean_fare) / std_fare

print(dataset[['Age_ZScore', 'Fare_ZScore',  'Age', 'Fare']])
```

```python
# Normalisasi Sigmoidal untuk kolom 'Age' dan 'Fare'
def sigmoidal_normalization(x):
    return (1 - np.exp(-x)) / (1 + np.exp(-x))

dataset['Age_Sigmoidal'] = sigmoidal_normalization(dataset['Age_ZScore'])
dataset['Fare_Sigmoidal'] = sigmoidal_normalization(dataset['Fare_ZScore'])

print(dataset[['Age_Sigmoidal', 'Fare_Sigmoidal',  'Age', 'Fare']])
```
