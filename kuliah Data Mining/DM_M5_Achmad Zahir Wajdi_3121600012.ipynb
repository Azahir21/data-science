```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold, LeaveOneOut, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# 1 dataset <-- titanic.csv
dataset = pd.read_csv('titanic.csv')
print(dataset.head())
```

```python
# 3 train_data <-- ambil dataset kolom fitur (Sex, Age, Pclass, Fare). Lakukan pengisian
# missing value pada fitur Age dengan nilai mean dari masing-masing class

dataset['Age'].fillna(dataset.groupby('Survived')['Age'].transform('mean'), inplace=True)
train_data = dataset.loc[:,['Sex', 'Age', 'Pclass', 'Fare']]
print(f"Data train: \n{train_data.head()}")
```

```python
# 4. label <-- ambil dataset kolom kelas (Survived)

label = dataset.loc[:, ['Survived']]
label = np.array(label).ravel()
print(f"label : \n{label}")
```

```python
# 5. train_data <-- lakukan normalisasi pada train_data dengan Min-Max 0-1 (catat nilai
# min dan max setiap atribut)

scaler = MinMaxScaler()
train_data[['Age', 'Pclass', 'Fare']] = scaler.fit_transform(train_data[['Age', 'Pclass', 'Fare']])
min_values = scaler.data_min_
max_values = scaler.data_max_

print(f'hasil normalisasi: \n{train_data.head()}\n')
print(f'nilai min setelah normalisasi: \n{min_values}')
print(f'nilai max setelah normalisasi: \n{max_values}')
```

```python
# 6. test_data <-- lakukan normalisasi pada train_data dengan min-max pada Langkah 5

def minmax(x, min, max):
    return (x - min) / (max - min)


test_data = minmax(train_data[['Age', 'Pclass', 'Fare']], min_values, max_values)

print(f'hasil normalisasi: \n{test_data.head()}\n')
```

```python
# 2a Hold-out Method (70%-30%)

train_data = pd.get_dummies(train_data, columns=['Sex'], drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(train_data, label, test_size=0.3, random_state=42)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_holdout = accuracy_score(y_test, y_pred)
```

```python
# 2b K-Fold (k=10)

kf = KFold(n_splits=10, shuffle=True, random_state=42)
scores_kfold = cross_val_score(knn, train_data, label, cv=kf)
accuracy_kfold = scores_kfold.mean()
```

```python
# 2c LOO

loo = LeaveOneOut()
scores_loo = cross_val_score(knn, train_data, label, cv=loo)
accuracy_loo = scores_loo.mean()
```

```python
# 7. Lakukan klasifikasi k-NN (k=3) untuk masing-masing pendekatan validasi dan
# hitunglah error ratio-nya

# Menghitung error ratio untuk setiap pendekatan
error_ratio_holdout = 1 - accuracy_holdout
error_ratio_kfold = 1 - accuracy_kfold
error_ratio_loo = 1 - accuracy_loo

# Menampilkan hasil error ratio
print("Error Ratio Hold-out (70%-30%):", error_ratio_holdout)
print("Error Ratio K-Fold (k=10):", error_ratio_kfold)
print("Error Ratio Leave-One-Out (LOO):", error_ratio_loo)
```
