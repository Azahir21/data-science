```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np

dataset = pd.read_csv('/content/drive/MyDrive/dataset/milk.csv') #pilih file csv
train_data = np.array(dataset)[:, 1:-1]
train_label = np.array(dataset)[:,-1]

kNN = KNeighborsClassifier(n_neighbors = 3, weights = 'distance')
kNN.fit(train_data, train_label)

angkaX = input('Tulis input X: ')
angkaY = input('Tulis input Y: ')
test_data = np.array([int(angkaX), int(angkaY)])
print(test_data)

test_data = np.reshape(test_data,(1, -1))
print(test_data)
hasil = kNN.predict(test_data)
print("Hasil dari k-NN : ", hasil)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np

dataset = pd.read_csv('/content/drive/MyDrive/dataset/milk_training.csv') #pilih file csv
train_data = np.array(dataset)[:, 1:-1]
train_label = np.array(dataset)[:,-1]

kNN = KNeighborsClassifier(n_neighbors = 3, weights = 'distance')
kNN.fit(train_data, train_label)

datatest = pd.read_csv('ruspini_testing.csv')
test_data = np.array(datatest)[:, 1:-1]
test_label = np.array(datatest)[:, -1]
hasil = kNN.predict(test_data)

print('Test Label: ', test_label)
print('Hasil k-NN :', hasil)
```
