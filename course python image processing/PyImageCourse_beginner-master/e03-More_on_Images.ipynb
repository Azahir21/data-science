# 3. More on images

1. Import the numpy and skimage packages

2. Using the skimage import function, load the image visible at this link: http://wiki.cmci.info/sampleimages/cells_Actin.tif
and put it in a variable called image

3. Check what are the dimensions of the image

4. Check what is the image type. Is it an 8 or 16 bit image ?

5. Plot the image using Matplotlib (that you have to import). Change the color map to 'gray' and make the image larger

6. Plot a histogram of the pixel values. Change the bin size to better see the distribution

```python
# 3.1. Import the numpy and skimage packages
import numpy as np
from skimage import io
import matplotlib.pyplot as plt

# 3.2. Load the image from the given URL and put it in a variable called image
image_url = "http://wiki.cmci.info/sampleimages/cells_Actin.tif"
image = io.imread(image_url)

# 3.3. Check the dimensions of the image
image_shape = image.shape
print("Image dimensions:", image_shape)

# 3.4. Check the image type (8 or 16 bit)
image_type = image.dtype
print("Image type:", image_type)

# 3.5. Plot the image using Matplotlib, change the colormap to 'gray', and make it larger
plt.figure(figsize=(8, 8))
plt.imshow(image, cmap='gray')
plt.axis('off')  # Turn off axis labels
plt.title("Original Image")
plt.show()

# 3.6. Plot a histogram of the pixel values with a custom bin size
plt.figure(figsize=(8, 4))
plt.hist(image.ravel(), bins=256, range=(0, 256), density=True, color='gray', alpha=0.7)
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("Histogram of Pixel Values")
plt.show()
```

# 3b. More on images: cropping, logical indexing

1. Make a cropped version of the image by selecting 100 rows for which you want all columns and plot it.

2. Select a region in the image of 100x100 pixesl and create a cropped image. Plot it.

3. Using this second cropped image, create a boolean mask by finding all pixels above some reasonable threshold (use the histogram above to find a value)

4. Plot your mask

5. Get all the values of all the pixels which are selected in the mask and store in a variable

6. Plot a histogram of all the values

```python
# 3b.1. Make a cropped version of the image by selecting 100 rows and all columns, then plot it
cropped_image_1 = image[:100, :]
plt.figure(figsize=(8, 8))
plt.imshow(cropped_image_1, cmap='gray')
plt.axis('off')
plt.title("Cropped Image (100 Rows)")
plt.show()

# 3b.2. Select a region in the image of 100x100 pixels and create a cropped image, then plot it
cropped_image_2 = image[100:200, 100:200]
plt.figure(figsize=(4, 4))
plt.imshow(cropped_image_2, cmap='gray')
plt.axis('off')
plt.title("100x100 Cropped Image")
plt.show()

# 3b.3. Create a boolean mask by finding all pixels above a threshold (you can choose a reasonable threshold)
threshold = 150
mask = cropped_image_2 > threshold

# 3b.4. Plot the mask
plt.figure(figsize=(4, 4))
plt.imshow(mask, cmap='gray')
plt.axis('off')
plt.title("Mask")
plt.show()

# 3b.5. Get all the values of the pixels selected in the mask and store in a variable
selected_pixel_values = cropped_image_2[mask]

# 3b.6. Plot a histogram of all the values in the mask
plt.figure(figsize=(8, 4))
plt.hist(selected_pixel_values.ravel(), bins=64, range=(0, 256), density=True, color='gray', alpha=0.7)
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("Histogram of Selected Pixel Values")
plt.show()
```
