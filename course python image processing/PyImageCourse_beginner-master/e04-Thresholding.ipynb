2. Import the same image from this url address: https://cildata.crbs.ucsd.edu/media/images/13585/13585.tif


3. Create a new variable called ```image_green``` that contains only the green channel by slicing the array

4. Plot this image

5. Plot a histogram of the image

6. Try to determine an appropriate threshold to recover the large structures in the image, create the corresponding mask and plot it

7. There are other algorithms for thresholding than the one of Otsu. Try to find one in the [Scikit-image documentation](http://scikit-image.org/docs/dev/api/skimage.filters.html) and to use it

````python
# 4.1. Import the relevant packages: numpy, matplotlib, and skimage
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, filters
from skimage.filters import threshold_yen

# 4.2. Import the image from the given URL
image_url = "https://cildata.crbs.ucsd.edu/media/images/13585/13585.tif"
image = io.imread(image_url)

# 4.3. Create a new variable called image_green that contains only the green channel by slicing the array
image_green = image[:, :, 1]  # Green channel is the second channel (0-based indexing)

# 4.4. Plot the green channel image
plt.figure(figsize=(6, 6))
plt.imshow(image_green, cmap='gray')
plt.axis('off')
plt.title("Green Channel Image")
plt.show()

# 4.5. Plot a histogram of the green channel image
plt.figure(figsize=(8, 4))
plt.hist(image_green.ravel(), bins=256, range=(0, 256), density=True, color='green', alpha=0.7)
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("Histogram of Green Channel Image")
plt.show()

# 4.6. Determine an appropriate threshold to recover large structures and create the corresponding mask
threshold = 60  # Adjust this threshold as needed
mask = image_green > threshold

# Plot the mask
plt.figure(figsize=(6, 6))
plt.imshow(mask, cmap='gray')
plt.axis('off')
plt.title("Thresholded Mask (Manual)")
plt.show()

# 4.7. Use the Yen thresholding method from scikit-image
yen_threshold = threshold_yen(image_green)
yen_mask = image_green > yen_threshold

# Plot the Yen thresholded mask
plt.figure(figsize=(6, 6))
plt.imshow(yen_mask, cmap='gray')
plt.axis('off')
plt.title("Thresholded Mask (Yen)")
plt.show()
````
