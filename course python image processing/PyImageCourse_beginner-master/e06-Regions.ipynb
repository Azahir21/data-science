# 6. Region properties exercises

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import skimage.io
import skimage.morphology
from skimage import color, measure
````

1. Import the image from http://wiki.cmci.info/sampleimages/4color_cells.tif

2. Create a mask using the third channel

3. Create a labeled image of the mask

4. Measure the region properties using this labeled image and use the **fourth** channel as intensity image

5. Extract the ```mean_intensity``` property from the dictionary

6. Plot a histogram of these intensities

7. What do you observe in this histogram ? Plot the fourth channel and see if the histogram makes sense.

````python
# 6.1. Import the image from the given URL
image_url = "http://wiki.cmci.info/sampleimages/4color_cells.tif"
image = skimage.io.imread(image_url)

# 6.2. Create a mask using the third channel (index 2, 0-based)
mask = image[:, :, 2]

# 6.3. Create a labeled image of the mask
labeled_image = skimage.morphology.label(mask)

# 6.4. Measure the region properties using the labeled image and use the fourth channel as the intensity image (index 3, 0-based)
intensity_image = image[:, :, 3]
region_properties = measure.regionprops(labeled_image, intensity_image)

# 6.5. Extract the mean_intensity property from the region properties
mean_intensities = [prop.mean_intensity for prop in region_properties]

# 6.6. Plot a histogram of these intensities
plt.hist(mean_intensities, bins=50, alpha=0.7, color='blue')
plt.xlabel("Mean Intensity")
plt.ylabel("Frequency")
plt.title("Histogram of Mean Intensities")
plt.show()

# 6.7. Plot the fourth channel to see if the histogram makes sense
plt.figure(figsize=(6, 6))
plt.imshow(intensity_image, cmap='gray')
plt.axis('off')
plt.title("Fourth Channel (Intensity Image)")
plt.show()
````
